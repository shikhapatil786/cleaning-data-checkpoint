```python
## to check how many null values are there in each column
df.isnull().sum()
```

```python
#what percentage of my data is null by column
df.isnull().sum() * 100 / len(df)
```

```python
#replace null value with unknown in director column
df['director']=df['director'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in cast column
df['cast']=df['cast'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in country column
df['country']=df['country'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in date_added column
df['date_added']=df['date_added'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in rating column
df['rating']=df['rating'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in duration column
df['duration']=df['duration'].fillna(value='unknown')
df.isnull().sum()
```

```python
#final check to see if there are any null values left in the dataset
df.isnull().sum()
```

```python
#duplicate rows in the dataset
df.duplicated().sum()
```
