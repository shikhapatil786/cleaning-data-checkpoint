```python
#import needed libraires with aliases
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('netflix_titles.csv')
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
#want to know the basic information about the dataset and to see if the dataset has imported propeerly
df.info(4)
```

```python
#want to have a look at the first 4 rows of the dataset
df.head(50)
```

```python
# want to know how many total data points are in the dataset
df.shape
```

```python
## to check how many null values are there in each column
df.isnull().sum()
```

```python
#what percentage of my data is null by column
df.isnull().sum() * 100 / len(df)
```

```python
#replace null value with unknown in director column
df['director']=df['director'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in cast column
df['cast']=df['cast'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in country column
df['country']=df['country'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in date_added column
df['date_added']=df['date_added'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in rating column
df['rating']=df['rating'].fillna(value='unknown')
df.isnull().sum()
```

```python
#replace null value with unknown in duration column
df['duration']=df['duration'].fillna(value='unknown')
df.isnull().sum()
```

```python
#final check to see if there are any null values left in the dataset
df.isnull().sum()
```

```python
#duplicate rows in the dataset
df.duplicated().sum()
```

## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
#Depending on if a column is numeric or categorical with determine what method to use for outliers 
# find the datatype of each column
df.dtypes
```

## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

```python
#this project i am only intersted in showing when the tv show or movies was added to netflix therefore i will drop the release year 
columns_to_drop = ['release_year','duration']
final_df=df.drop(columns_to_drop, axis=1)
final_df.info(4)
```

```python
# define the file name 
filename = 'Netflix_final_dataset.csv'
```

```python
#saving the dataframe as csv file
final_df.to_csv(filename, index=False)
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
#looking for inconsistent data for first 70 records from dataset.
final_df.head(70)
```

```python
#looking for inconsistent data from dataset.
final_df['country']=final_df['country'].str.strip().str.title()
```

```python
#trying to split the country column into multiple rows to get a better understanding of the data and 
# to be able to plot the top countries for better visualization
final_df=final_df.explode('country')
print(final_df['country'].value_counts())

```

```python
#looking for inconsistent data from dataset.
final_df['date_added'].value_counts()
```

```python
#counting the number of unique type  to see if there is any inconsistency in the type column and 
# to see how many movies and TV shows are in the dataset. 
# This will help us to understand the distribution of movies and TV shows in the dataset and 
# to see if there are any outliers or inconsistencies in the type column.
final_df['type'].value_counts()
```

```python
#counting the number of unique show_id in the dataset to see if there are any duplicate show_ids in the dataset. 
# If there are duplicate show_id, it could indicate that there are duplicate rows in the dataset, 
# which could affect the analysis and visualization of the data. By counting the number of unique show_id, 
# we can get an idea of how many unique movies and TV shows are in the dataset and if there are any duplicates 
# that need to be addressed.
final_df['show_id'].value_counts()
```

```python
#counting the number of unique ratings in the dataset to see if there are any inconsistencies in the rating 
#column and to see how many movies and TV shows are in the dataset with each rating.
final_df['rating'].value_counts()
```

```python
#data shows signs that multiple production from different countries collaborated on a single movie/title
#sort by count and plot the top countries for better visualization
country_counts = final_df['country'].value_counts().head(25)
country_counts.plot(kind='barh', figsize=(10, 8))
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.title('Top 25 Countries by Number of Titles')
plt.tight_layout()
plt.show()
```
